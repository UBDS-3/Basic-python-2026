### Exercise 12. Count vowels and consonants in a sentence / Вправа 12. Підрахуйте голосних і приголосних у реченні
Write a program that counts the total number of vowels and consonants in a given sentence.

Напишіть програму, яка підраховує загальну кількість голосних та приголосних у заданому реченні.

**Expected Output / Очікуваний результат:**
```
Vowels: 5
Consonants: 7
```

````python
# Your solution here / Ваше рішення тут
sentence = "Myxobacteria produce secondary metabolites"

vowels = "aeiouAEIOU"

vowel_count = 0
consonant_count = 0

for char in sentence:
    if char.isalpha():
        if char in vowels:
            vowel_count = vowel_count + 1
        else:
            consonant_count = consonant_count + 1

print("Sentence:", sentence)
print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
````

### Exercise 15. Find largest and smallest digit in a number / Вправа 15. Знайдіть найбільшу та найменшу цифру в числі
Write a program to find the largest and smallest digit within a given integer.

Напишіть програму для знаходження найбільшої та найменшої цифри у заданому цілому числі.

**Expected Output / Очікуваний результат:**
```
Largest: 9, Smallest: 5
```

````python
# Your solution here / Ваше рішення тут
num = 583921

n = num
largest = -1
smallest = 10

while n > 0:
    digit = n % 10

    if digit > largest:
        largest = digit
    if digit < smallest:
        smallest = digit

    n = n // 10

print("Number:", num)
print("Largest digit:", largest)
print("Smallest digit:", smallest)
````
